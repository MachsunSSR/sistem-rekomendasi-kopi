```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# import data
df = pd.read_csv('datamuz3.csv')
df.head()
```

```python
# Remove column Timestamp, Email Address, Nama, Jenis kelamin, Umur
df = df.drop(['Timestamp', 'Email Address', 'Nama', 'Jenis kelamin', 'Umur'], axis=1)

# Create 1 new column that consist of value in Coffee Based / Kopi, Chocolate, Soda, Tea, and Milk Based column and Remove the Nan
df['Preferensi'] = df[['Coffee Based / Kopi', 'Chocolate', 'Soda', 'Tea', 'Milk Based']].values.tolist()
df['Preferensi'] = df['Preferensi'].apply(lambda x: [i for i in x if str(i) != 'nan'])

# Remove Coffee Based / Kopi, Chocolate, Soda, Tea, and Milk Based column
df = df.drop(['Coffee Based / Kopi', 'Chocolate', 'Soda', 'Tea', 'Milk Based'], axis=1)
```

```python
df.head()
```

```python
# Change first column name to "Deskripsi"
df.rename(columns={'Deskripsikan minuman yang anda inginkan\nContoh : ( Saya menginginkan minuman yang berbahan dasar dari kopi tetapi saya inginnya yang manis dan dingin )': 'Deskripsi'}, inplace=True)
```

```python
df.head()
```

```python
#Remove indonesian Stopwords
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import string


stop_words = set(stopwords.words('indonesian'))
factory = StemmerFactory()

# Create new column that consist of stemmed words from Deskripsi column
df['Cleaned'] = df['Deskripsi'].apply(lambda x: ' '.join([factory.create_stemmer().stem(word.lower()) for word in x.split()]))

# Remove Stopwords
df['Cleaned'] = df['Cleaned'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))

df['Cleaned'] = df['Cleaned'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))

# Remove number and all symbols
df['Cleaned'] = df['Cleaned'].str.replace('\d+', '')

df.head()
```

```python
# Remove row that column preferensi contain ','
df = df[~df['Preferensi'].astype(str).str.contains(',')]
df.head()
```

```python
# Create cleaning function same as above, but is used for cleaning the text, not for dataframe

def cleaning(text):
    stop_words = set(stopwords.words('indonesian'))
    factory = StemmerFactory()
    text = ' '.join([factory.create_stemmer().stem(word.lower()) for word in text.split()])
    text = ' '.join([word for word in text.split() if word not in (stop_words)])
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = text.replace('\d+', '')
    return text
```

```python
# Save the cleaned data to csv
df.to_csv('cleaned3.csv', index=False)
```

```python
# Use TF-IDF to vectorize the words in Cleaned columns
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
X = tfidf.fit_transform(df["Cleaned"])
y = df["Preferensi"]

# Import Random Forest
from sklearn.ensemble import RandomForestClassifier

# Train the X
model = RandomForestClassifier()
model.fit(X, y)
```

```python
# Hyperparameter Tuning
from sklearn.model_selection import GridSearchCV

param_grid = {
    'bootstrap': [True],
    'max_depth': [80, 90, 100],
    'max_features': [2, 3],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [8, 10, 12],
    'n_estimators': [100, 200, 300, 1000]
}

grid_search = GridSearchCV(estimator = model, param_grid = param_grid,
                            cv = 3, n_jobs = -1, verbose = 2)

grid_search.fit(X, y)
```

```python
# Save the model
import pickle
pickle.dump(model, open('model.pkl','wb'))
```

```python
# Save the TF-IDF
pickle.dump(tfidf, open('tfidf.pkl','wb'))
```

```python
# import model
model = pickle.load(open('model.pkl','rb'))

# import tfidf
tfidf = pickle.load(open('tfidf.pkl','rb'))

# Try Testintg
text = ["coklat manis"]
text = tfidf.transform(text)

predicted_probabilities = model.predict_proba(text)
classes = model.classes_

# Get top 5 classes
top_5 = np.argsort(predicted_probabilities, axis=1)[:, -5:]
top_5 = np.flip(top_5, axis=1)

# Print top 5 classes
print("Top 5 classes:")
for row in top_5:
    for idx in row:
        print(classes[idx], end=" ")
    print()
```
